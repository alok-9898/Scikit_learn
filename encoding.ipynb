```python
#Feature Encoding
#numerical datas are scaled using the preprocessing but some datas that are not numerical are to be encoded using feature encoding
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car', as_frame=True).frame
```

```python
data
```

```python
columns_to_encode = ['lug_boot', 'safety']

encoder = OrdinalEncoder(
    categories = [
        ['small', 'med', 'big'],
        ['low', 'med', 'high']
    ]
)
data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
#Now the datas are encoded and changed into some numerical values
data
```

```python
#reverse encoding the data
encoder.inverse_transform(data[columns_to_encode])
```
